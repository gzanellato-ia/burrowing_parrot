## Algorithm to create empty patches, 256 x 256
To use as target in images that do not have nest entrances

```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Creating the empty dots map

```python
# Specifying array as a tuple, and specify their data types.
ceros = np.zeros((256, 256, 4), dtype='i8')
ceros.shape
```

```python
empty_patch = ceros + [0,0,0,255]
empty_patch
```

```python
io.imshow(empty_patch)
```

```python
cv2.imwrite('black_patch.png', empty_patch)
```

```python
cant_actual = 5
ultima = 160    # IMG_4389_160.png

for i in range(cant_actual):
    
    # saving the image
    a = ('0' + str(i + ultima + 1))[-3::]
    nombre_i = ('empty_patches_dots/IMG_4389_' + a + '.png')
    #print(nombre_i)
    
    io.imsave(nombre_i, empty_patch)
```
